### Применить фреймворк ICE для приоритизации гипотез. Отсортировать их по убыванию приоритета.

Сформируем ICE

```python
hypothesis['ICE'] = hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']
```

```python
hypothesis_ICE = hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False).round(2)
```

```python
hypothesis_ICE
```

### Применить фреймворк RICE для приоритизации гипотез. Отсортировать их по убыванию приоритета.

Сформируем RICE

```python
hypothesis['RICE'] = hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']
```

```python
hypothesis_RICE = hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
```

```python
hypothesis_RICE
```

### Указать, как изменилась приоритизация гипотез при применении RICE вместо ICE. Объяснить, почему так произошло.

Параметр охвата reach изменил приоретизацию гипотез при методе ICE. Объеденим таблицы

```python
priority_of_hypothesis = hypothesis_ICE.merge(hypothesis_RICE, on='hypothesis')
```

```python
priority_of_hypothesis
```

## Часть. Анализ A/B-теста

### Обработка и подготовка данных

```python
orders = pd.read_csv('/datasets/orders.csv')
visitors = pd.read_csv('/datasets/visitors.csv')
```

```python
orders
```

```python
orders.info()
```

```python
visitors
```

```python
visitors.info()
```

ПРиведем столбец date к формату datetime

```python
visitors['date'] = pd.to_datetime(visitors['date'], format = '%Y-%m-%d %H:%M:%S')
```

Создадим массив уникальных пар

```python
datesGroups = orders[['date','group']].drop_duplicates() 
```

Агрегируем значения.

```python
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]
.agg({'date' : 'max', 'group' : 'max', 'transactionId' : pd.Series.nunique, 'visitorId' : pd.Series.nunique, 'revenue' : 'sum'}), 
axis=1).sort_values(by=['date','group'])
```

```python
visitorsAggregated = datesGroups.apply(lambda x: 
visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]
.agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])
```

```python
ordersAggregated['date'] = pd.to_datetime(ordersAggregated['date'], format = '%Y-%m-%d %H:%M:%S')
```

Мерджим таблицы

```python
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
```

```python
cumulativeData
```

### Построить график кумулятивной выручки по группам. Сделать выводы и предположения.

Дата с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А

```python
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
```

Дата с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B

```python
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]
```

```python
from pandas.plotting import register_matplotlib_converters
plt.figure(figsize=(12,4))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')
plt.title('Кумулятивная выручка в группах А и В по дням')
plt.xlabel('Дата', )
plt.ylabel('Кумулятивная выручка')
plt.legend();
```

Вывод: Выручка повышается в обеих группах во время всего теста. Выручка группы В больше выручка группы А. Выручка в группе В резко возрастает в 1 точке. Это может свидетельствовать либо о всплексе количества заказов либо о появлении очень дорогих заказов.

### Построить график кумулятивного среднего чека по группам. Сделать выводы и предположения.

```python
plt.figure(figsize=(12,4))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue']/cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue']/cumulativeRevenueB['orders'], label='B')
plt.title('Кумулятивная средний чек в группах А и В по дням')
plt.xlabel('Дата')
plt.ylabel('Средний чек')
plt.legend() 
```

Вывод: Средний чек в группе В тоже в общем выше чем в группе А и в обеих группах он увеличивается. В группе В так же наблюдается скачок.

### Построить график относительного изменения кумулятивного среднего чека группы B к группе A. Сделать выводы и предположения.

Cобираем данные в одном датафрейме

```python
mergedCumulativeRevenue = cumulativeRevenueA.merge(cumulativeRevenueB, left_on='date', right_on='date', how='left', suffixes=['A', 'B'])
```

Cтроим отношение средних чеков

```python
plt.figure(figsize=(12,4))
plt.plot(mergedCumulativeRevenue['date'], (mergedCumulativeRevenue['revenueB']/mergedCumulativeRevenue['ordersB'])/(mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA'])-1)
plt.title('Относительное изменение кумулятивного среднего чека группы В к группе А')
plt.xlabel('Дата')
plt.ylabel('Средний чек')
plt.axhline(y=0, color='black', linestyle='--') 
```

Вывод: Отношение тоже растет. Но в некоторых точках заметны скачки. Видимо точно есть крупные заказы и выбросы

### Построить график кумулятивной конверсии по группам. Сделать выводы и предположения.

Считаем коммулятивную конверсию

```python
cumulativeData['conversion'] = cumulativeData['orders']/cumulativeData['visitors']
```

Отделим данные по группам

```python
cumulativeDataA = cumulativeData[cumulativeData['group']=='A']
```

```python
cumulativeDataB = cumulativeData[cumulativeData['group']=='B']
```

### Посчитать статистическую значимость различий в конверсии между группами по «сырым» данным. Сделать выводы и предположения. 

Сформируем нулевую гипотезу. Н0 = Коверсии между группами по "сырым" данным не отличается.

Альтернативная гипотеза Н1 = Коверсии между группами по "сырым" данным отличаются.

```python
ordersByUsersA = orders[orders['group']=='A'].groupby('visitorId', as_index=False).agg({'transactionId':'nunique'})
ordersByUsersA.columns = ['visitorId', 'orders']
ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False).agg({'transactionId':'nunique'})
ordersByUsersB.columns = ['visitorId', 'orders']
```

Cоставим списки кол-ва заказов sampleA и sampleB со всеми пользователями (в том числе и не купившими) по группам

```python
list_orders_1 =[]
for i in range(0,(visitors[visitors['group']=='A']['visitors'].sum()-len(ordersByUsersA))):
    list_orders_1.append(0)
orders_by_non_purchased_users_A = pd.Series(data = list_orders_1, name = 'orders')

list_orders_2 =[]
for i in range(0,(visitors[visitors['group']=='B']['visitors'].sum()-len(ordersByUsersB))):
    list_orders_2.append(0)
orders_by_non_purchased_users_B = pd.Series(data = list_orders_2, name = 'orders')

sampleA = pd.concat([ordersByUsersA['orders'], orders_by_non_purchased_users_A], axis=0)
sampleB = pd.concat([ordersByUsersB['orders'], orders_by_non_purchased_users_B], axis=0)
```

```python
print('\nОтносительный прирост конверсии группы B по отношению к группе A: {:.3f}'.format(sampleB.mean()/sampleA.mean()-1))
```

```python
alpha = 0.05
results = stats.mannwhitneyu(sampleA, sampleB, alternative = 'two-sided')
print('P-value: {}'.format(results.pvalue))
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Вывод: конверсии по "сырым" данным между группами отличаются

### Посчитать статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным. Сделать выводы и предположения.

Сформируем нулевую гипотезу. Н0 = Различий в среднем чеке между группами по "сырым" данным нет.

Альтернативная гипотеза Н1 = Различия в среднем чеке между группами по "сырым" данным есть.

### Посчитаь статистическую значимость различий в конверсии между группами по «очищенным» данным. Сделать выводы и предположения.

Ранее мы уже выяснили, что подавляющее большинство пользователей не делает более 2 заказов и сумма 1 покупки в большинстве случаев не превышает 28 т.р. Очистим данные от аномальных пользователей с превышающими данные лимиты значениями.

```python
limit_orders = 2
limit_revenue = 28000
```

```python
usersWithManyOrders = pd.concat([ordersByUsersA[ordersByUsersA['orders']>limit_orders]['visitorId'], 
                                ordersByUsersB[ordersByUsersB['orders']>limit_orders]['visitorId']], axis = 0)
usersWithExpensiveOrders = orders[orders['revenue']>limit_revenue]['visitorId']
abnormalUsers = pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis = 0).drop_duplicates().sort_values()
print('Кол-во аномальных пользователей: {}'.format(len(abnormalUsers)))
```

Сформируем нулевую гипотезу. Н0 = Коверсии между группами по "очищенным" данным не отличается.

Альтернативная гипотеза Н1 = Коверсии между группами по "очищенным" данным отличаются.

```python
sampleAFiltered = pd.concat([
    ordersByUsersA[np.logical_not(ordersByUsersA['visitorId'].isin(abnormalUsers))]['orders'],
    orders_by_non_purchased_users_A
], axis = 0)
sampleBFiltered = pd.concat([
    ordersByUsersB[np.logical_not(ordersByUsersB['visitorId'].isin(abnormalUsers))]['orders'],
    orders_by_non_purchased_users_B
], axis = 0)
print('\nОтносительный прирост конверсии группы B по отношению к группе A после очистки данных: {:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
results = stats.mannwhitneyu(sampleAFiltered, sampleBFiltered, alternative = 'two-sided')
print('P-value: {}'.format(results.pvalue))
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Вывод: Коверсия между группами по "очищенным" данным отличается.

Сформируем нулевую гипотезу. Н0 = Различий в среднем чеке между группами по "очищенным" данным нет.
